```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load 
border_data_to_load = "Border_Crossing_Entry_Data.csv"

# Read the Border Corssing Entry Data file and store into a Pandas Data Frame
border_data_df = pd.read_csv(border_data_to_load)
```

```python
# Confirm full file was loaded
len(border_data_df)
```

```python
# Preview the data frame
border_data_df.head(15)
```

```python
# Convert 'Date' from str to datetime format and store in a new field called 'Date_dt'
border_data_df ['Date_dt'] = pd.to_datetime(border_data_df['Date'])
```

```python
# Preview the data frame
border_data_df.head(15)
```

```python
# Create a copy of border_data_df to use for the president analytics
border_pres_df = border_data_df
```

```python
# Create a column to contain the year of each record
border_pres_df['Year'] = pd.DatetimeIndex(border_pres_df['Date_dt']).year
```

```python
# Filter now the data to only include 2002 - 2015 (Bush's terms and Obama's terms)
border_pres_df = border_pres_df.drop(border_pres_df[border_pres_df.Year < 2001].index)
border_pres_df = border_pres_df.drop(border_pres_df[border_pres_df.Year > 2016].index)
len(border_pres_df)
```

```python
border_pres_df.head()
```

```python
# For each record, determine the president in office
border_pres_df['Pres'] = np.where(border_pres_df['Year']< 2009, 'Bush', 'Obama')
border_pres_df.head()
```

```python
# Print a bar chart that shows the total border crossings per year and president
border_pres_summary = border_pres_df.groupby(['Pres','Year'])['Value'].sum()

p=border_pres_summary.plot.bar(title="Border crossing over the presidential terms")
p.hlines(border_pres_summary.mean(), 0, 15, alpha=1)
p.grid()
```
